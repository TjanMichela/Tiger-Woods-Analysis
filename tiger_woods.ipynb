```python
import pandas as pd
import numpy as np
```

```python
# SG:Total dataframes
total_2005 = pd.read_csv("data/2005_total.csv")
total_2006 = pd.read_csv("data/2006_total.csv")
total_2007 = pd.read_csv("data/2007_total.csv")
total_2009 = pd.read_csv("data/2009_total.csv")

total_2005.head()
```

```python
# Tiger's SG:Total Average
tiger_total_2005 = total_2005[total_2005["PLAYER"] == "Tiger Woods"].AVG
tiger_total_2006 = total_2006[total_2006["PLAYER"] == "Tiger Woods"].AVG
tiger_total_2007 = total_2007[total_2007["PLAYER"] == "Tiger Woods"].AVG
tiger_total_2009 = total_2009[total_2009["PLAYER"] == "Tiger Woods"].AVG

tiger_total = (tiger_total_2005.add(tiger_total_2006, tiger_total_2007, tiger_total_2009))/4
tiger_total = round(tiger_total.values[0],3)

tiger_total
```

```python
# Top 50 SG:Total average
sg_total_2005 = np.average(total_2005["AVG"].head(50))
sg_total_2006 = np.average(total_2006["AVG"].head(50))
sg_total_2007 = np.average(total_2007["AVG"].head(50))
sg_total_2009 = np.average(total_2009["AVG"].head(50))

sg_total = round((sg_total_2005 + sg_total_2006 + sg_total_2007 + sg_total_2009)/4, 3)

sg_total
```

```python
stat_averages.to_csv("data/stat_averages.csv")
```

```python
tiger_woods_tournament_history = pd.read_csv("data/tiger_woods_tournament_history.csv")

tiger_woods_tournament_history.head(10)
```

```python
import addfips

state_fips = []

af = addfips.AddFIPS()
for state in tiger_woods_tournament_history["State"]:
  state_fips.append(af.get_state_fips(state))

tiger_woods_tournament_history["state_fips"] = state_fips   
tiger_woods_tournament_history.head() 
```

```python
heatmap_data = tiger_woods_tournament_history[["State", "state_fips"]].dropna()
heatmap_data["count"]=heatmap_data.groupby("State")['State'].transform("count")
heatmap_data.drop_duplicates(inplace=True)
heatmap_data
```

```python
heatmap_data.to_csv("data/heatmap_data.csv")
```
